# Ignorieren des Vorzeichens des Filterergebnisses

```python
abs_M = abs(M)
```

```python
abs_M.min()
```

```python
abs_M.max()
```

```python
fig = plt.figure(figsize=(20,10))
plt.imshow(abs_M, cmap="gray")
plt.colorbar()
plt.show()
```
